```python

import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from google.colab import drive
drive.mount("/content/drive")

df = pd.read_csv("/content/drive/MyDrive/hiring.csv")
```

```python
#Previewed the first 5 rows of the dataset
df.head()
```

```python
#Filled missing values in the experience column
df.experience = df.experience.fillna("zero")

#Converted text values in the experience column to numeric values
df.experience = df.experience.replace({"zero":0, "two":2, "three":3, "five":5, "seven":7,"ten":10, "eleven":11})
```

```python
#Filled missing values in the test score column using the median value
df["test_score(out of 10)"] = df["test_score(out of 10)"].fillna(df["test_score(out of 10)"].median())
```

```python
#Checked the data types of each column
df.dtypes
```

```python
#Separated input features (independent variables)
x = df.drop("salary($)", axis=1)

#Defined target variable (dependent variable)
y = df["salary($)"]


#Created a Linear Regression model
reg = LinearRegression()

#Trained the model using the dataset
reg.fit(x,y);
```

```python
#Predicted salary for candidate with:
#2 years experience, 9 test score, 6 interview score

round(reg.predict([[2,9,6]])[0], 2)
```

```python
#Predicted salary for candidate with:
#12 years experience, 10 test score, 10 interview score
round(reg.predict([[12,10,10]])[0], 2)
```
